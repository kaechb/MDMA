```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns

def plot_corr( data, leg=-1):



    correlation_matrix =   np.corrcoef(data, rowvar=False)

    np.fill_diagonal(correlation_matrix, np.nan)


    fig, ax = plt.subplots(1, 1, figsize=(8, 6.4))
    sns.heatmap(correlation_matrix,ax=ax, cmap='coolwarm', cbar=False)

    ax.set_xticks([0.5,9.5,19.5,29.5],labels=[1,10,20,30], rotation=0,fontsize=15)
    ax.set_yticks([0.5,9.5,19.5,29.5],labels=[1,10,20,30],rotation=0,fontsize=15)
    ax.set_xlabel("Particles",fontsize=20)
    ax.set_ylabel("Particles",fontsize=20)
    #cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
    plt.tight_layout()
    fig.colorbar(ax.collections[0])


    plt.show()

```

```python

x=torch.randn(1000000,30)
x=x.exponential_(lambd=1)
# x+=x.min(dim=1)[0].unsqueeze(1)[0]
x=x.sort(dim=1,descending=True)[0]
x=x/x.sum(dim=1 ).unsqueeze(1)
plot_corr(x.numpy())
```

```python

x=torch.randn(1000000,30)
xbar=x.mean(dim=1).unsqueeze(1)
x=x-xbar
# x+=x.min(dim=1)[0].unsqueeze(1)[0]
x=x.sort(dim=1,descending=True)[0]

plot_corr(x.numpy())
```
