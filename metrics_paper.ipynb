```python

import pandas as pd
best_df=pd.read_csv("results/best.csv")
in_df=pd.read_csv("results/in.csv")
epic_df=pd.read_csv("results/epic.csv")
# best_df["w1m"]=best_df["w1m_2"]
# epic_df["w1m"]=epic_df["w1m_2"]
# in_df["w1m"]=in_df["w1m_2"]
```

```python
import ast
cols=[]
for col in best_df.columns:
        cols.append(col)
print_table_epic=epic_df.copy()
print_table_epic["model"]="EPiC"
print_table=best_df[cols].copy()
print_table["model"]="MF"

print_table_in=in_df.copy()
print_table_in["model"]="IN"
print_table=pd.concat([print_table_epic,print_table,print_table_in],axis=0)
print_table=print_table.set_index(["name","model"],drop=False)
print_table[cols]
```

```python
import numpy as np
for col in ["w1m","w1p","w1efp","kpd","fpd"]:

        print_table[col]=print_table[col].str.replace("array","").apply(ast.literal_eval)

from tkinter import W
import pandas as pd
from numpy import array
import ast

def format_mean_sd(mean, sd,efp=False):
    """round mean and standard deviation to most significant digit of sd and apply latex formatting"""
    decimals = -int(np.floor(np.log10(sd)))+int(efp)+1
    decimals -= int((sd * 10 ** decimals) >= 9.5)
    if decimals < 0:
        ten_to = 10 ** (-decimals)
        if mean > ten_to:
            mean = ten_to * (mean // ten_to)
        else:
            mean_ten_to = 10 ** np.floor(np.log10(mean))
            mean = mean_ten_to * (mean // mean_ten_to)
        sd = ten_to * (sd // ten_to)
        decimals = 0

    if mean >= 1e3 and sd >= 1e3:
        mean = np.round(mean * 1e-3)
        sd = np.round(sd * 1e-3)
        return f"${mean:.{decimals}f}$k $\\pm {sd:.{decimals}f}$k"
   
    else:
        return f"${mean:.{decimals}f} \\pm {sd:.{decimals}f}$"

# print_table.index=["t"]#"q"

def weighted_mean(x):
    x,w=np.array(x[0]),np.array(x[1]) 
    weights=1/w**2
    return np.sum(x*weights)/np.sum(weights)
def weighted_std(x):
    x,w=np.array(x[1]),np.array(x[1])
    weights=1/w**2
    return np.sum(x*weights)/np.sum(weights)

print_table["pmm"]=print_table["w1m"].apply(lambda x:x).apply(weighted_std)
print_table["w1m"]=print_table["w1m"].apply(lambda x:x).apply(weighted_mean)
print_table["pmp"]=print_table["w1p"].apply(lambda x:x).apply(weighted_std)
print_table["w1p"]=print_table["w1p"].apply(lambda x:x).apply(weighted_mean)
print_table["pme"]=print_table["w1efp"].apply(lambda x:x).apply(weighted_std)
print_table["w1efp"]=print_table["w1efp"].apply(lambda x:x).apply(weighted_mean)
print(print_table["pme"])
print_table["fpd_std"]=print_table["fpd"].apply(lambda x:x[1]).apply(np.mean)
print_table["fpd"]=print_table["fpd"].apply(lambda x:x[0]).apply(np.mean)
print_table["kpd_std"]=print_table["kpd"].apply(lambda x:x[1]).apply(np.mean)
print_table["kpd"]=print_table["kpd"].apply(lambda x:x[0]).apply(np.mean)

cols=["w1m","w1efp","w1m","pmm","pme","pmp"]


# print_table=print_table[["w1m","w1p","w1efp","fpnd","cov","mmd","pmm","pmp","pme","model"]]

print_table.loc[:,"w1m"]*=1000
print_table.loc[:,"w1p"]*=1000
print_table.loc[:,"w1efp"]*=100000
print_table.loc[:,"pmm"]*=1000
print_table.loc[:,"pmp"]*=1000
print_table.loc[:,"pme"]*=100000
print_table.loc[:,"fpd"]*=10000
print_table.loc[:,"kpd"]*=10000
print_table.loc[:,"fpd_std"]*=10000
print_table.loc[:,"kpd_std"]*=10000


index=["q","g","t","w","z"]

print_table.loc[:,"w1m"]=print_table.apply(lambda x:format_mean_sd(float(x["w1m"]),float(x["pmm"])),axis=1)
print_table.loc[:,"w1efp"]=print_table.apply(lambda x:format_mean_sd(float(x["w1efp"]),float(x["pme"]),efp=True),axis=1)
print_table.loc[:,"w1p"]=print_table.apply(lambda x:format_mean_sd(float(x["w1p"]),float(x["pmp"])),axis=1)
print_table.loc[:,"kpd"]=print_table.apply(lambda x:format_mean_sd(float(x["kpd"]),float(x["kpd_std"])),axis=1)
print_table.loc[:,"fpd"]=print_table.apply(lambda x:format_mean_sd(float(x["fpd"]),float(x["fpd_std"])),axis=1)


# print_table.loc[:,"cov"]=print_table.loc[:,"cov"].astype(float).map('{:.2f}'.format)

# print_table.loc[:,"mmd"]=print_table.loc[:,"mmd"].astype(float).map('{:.3f}'.format)
print_table.loc[:,"pmm"]=print_table.loc[:,"pmm"].astype(float).map('{:,.2f}'.format)

print_table.loc[:,"pmp"]=print_table.loc[:,"pmp"].astype(float).map('{:,.2f}'.format)

print_table.loc[:,"pme"]=print_table.loc[:,"pme"].astype(float).map('{:,.2f}'.format)

# print_table.loc[:,"w1m"]="$"+print_table["w1m"].map(str)+"\pm"+print_table["pmm"].map(str)+"$"
# print_table.loc[:,"w1p"]="$"+print_table["w1p"].map(str)+"\pm"+print_table["pmp"].map(str)+"$"
# print_table.loc[:,"w1efp"]="$"+print_table["w1efp"].map(str)+"\pm"+print_table["pme"].map(str)+"$"
# print_table.loc[:,"kpd"]="$"+print_table["kpd"].map(str)+"\pm"+print_table["kpd_std"].map(str)+"$"
# print_table.loc[:,"fpd"]="$"+print_table["fpd"].map(str)+"\pm"+print_table["fpd_std"].map(str)+"$"

print_table=print_table.loc[index,:]
print_table
```

```python


tex=r"""\begin{table}[]
    \centering
    \begin{tabular}{ll|lllll}Jet Class & Model & $W_1^M (\times 10^{3})$ & $W_1^P
(\times 10^{3})$ & $W_1^{EFP}(\times 10^{5})$ &KPD (\times 10^{4})&FPD(\times 10^{4})\\\cline{1-7}
\cline{1-7}"""
for p in ["q","g","t","w","z"]:
    temp_full=print_table[print_table["name"]==p].copy()
    temp=temp_full[temp_full["model"]!="IN"]

    for col in ["w1m","w1p","w1efp","kpd","fpd"]:
        



        temp_vals=temp.loc[temp["model"]!="IN",col].astype(str).str.replace("$","").str.split("\\").str[0].astype(float)
        min_unc=temp.loc[temp["model"]!="IN",col].astype(str).str.replace("$","").str.split("pm").str[1].astype(float).min()
        min_range = (temp_vals.min() - min_unc, temp_vals.min() + min_unc)
        
        compatible = (temp_vals >= min_range[0]) & (temp_vals <= min_range[1])
        compatible[temp_vals.argmin()]=False
        if not compatible.any():
            temp_full.loc[:,col].values[temp_vals.argmin()]="$\mathbf{"+temp_full.loc[:,col].values[temp_vals.argmin()].replace("$","")+"}$"

    temp=temp_full[["model","w1m","w1p","w1efp","kpd","fpd"]]

    temp.columns=["model","$W_1^M (\times 10^{3})$","$W_1^P (\times 10^{3})$","$W_1^{EFP}(\times 10^{5})$","FPD","KPD"]
    text=temp.to_latex(index=False,escape=False)
    parton="Gluon" if p=="g" else "Light Quark" if p=="q" else "Top Quark" if p=="t" else "W" if p=="w" else "Z"
    k=1 if p in ["w","z"] else 2
    tex+="\multirow{"+str(k)+"}{*}{"+parton+"} & "+text.split("KPD \\\\")[1].split("\\bottomrule")[0].replace("\\\\","\\\\&").replace("\\midrule","").replace("MP_LFC","MP\_LFC").replace("  ","")[:-2]+"\cline{1-7}" 
    tex+="\n"


tex+="""  \end{tabular}

\end{table}"""
print(tex)
```
